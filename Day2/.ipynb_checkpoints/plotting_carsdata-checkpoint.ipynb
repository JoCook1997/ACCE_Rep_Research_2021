```python
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
#specify directory where data is stored, so any following paths are relative to this directory
os.chdir(r'C:/Users/JoCook/Documents/PhD/ACCE_Reproducible_Research_Course/Day2/')

#read in data
data=pd.read_csv(r'Cars.csv')
```

```python
#simple scatter plot with different coloured points based on number of cyclinders
ax=sns.scatterplot(x=data['Horsepower'], y=data['MPG_City'], data=data, hue=data['Cylinders'])
```

```python
%%time

#Unable to figure out how to get a key for cylinders
#Here I change font sizes, backgrounds, 
#gridlines, tick frequencies and fonts

#plot horsepower on x and MPG_City on y

#------------------------------------------------------

#set up font dictionary
#set up the font dictionaries
font = {'family': 'Arial',
        'color':  'black',
        'size': 18,
        } 
#------------------------------------------------------

#set up the plot
fig, ax1 = plt.subplots(figsize=(10,5))

#change the background colour of the plot 
ax1.set_facecolor('lightgray')

#label the axes
ax1.set_ylabel('MPG_City',fontdict=font)
ax1.set_xlabel('Horsepower',fontdict=font)

#title the plot
plt.title('Scatter plot between MPG_City and Horsepower',fontdict=font)

#edit which spines are visible
ax1.spines['top'].set_visible(True)
ax1.spines['right'].set_visible(True)
ax1.spines['bottom'].set_visible(True)
ax1.spines['left'].set_visible(True)

#-------------------------------------------------------

#set up a list of colour labels to colour the points 
#differently based on number of cylinders
color_labels = data['Cylinders'].unique()
rgb_values = sns.color_palette("Set2", len(color_labels))
color_map = dict(zip(color_labels, rgb_values))

#-------------------------------------------------------

#plot the data
ax1.scatter(data['Horsepower'], data['MPG_City'],c=data['Cylinders'].map(color_map))

#------------------------------------------------------
#format ticks

#tick frequency
ax1.set_yticks(np.arange(data['MPG_City'].min(), data['MPG_City'].max()+1, step=10.0))

#tick fontsize
for tick in ax1.get_xticklabels():
       tick.set_fontname("Arial"); tick.set_fontsize(14);
for tick in ax1.get_yticklabels():
       tick.set_fontname("Arial"); tick.set_fontsize(14);
ax1.tick_params('x',rotation=0, colors='black')
ax1.tick_params('y',colors='black')

#------------------------------------------------------
#set up gridlines
ax1.set_axisbelow(True)
ax1.grid(b=True, which='major', linestyle='-', linewidth='0.5', color='white')

#------------------------------------------------------

plt.show()
```
